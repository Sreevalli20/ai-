```python
# ===============================
# INSTALL REQUIREMENTS
# ===============================
!pip install streamlit pyngrok -q

from pyngrok import ngrok
import threading
import os

# ===============================
# ADD YOUR NGROK TOKEN
# ===============================
!ngrok config add-authtoken 3BLRiJr0z3Di5d0vqARFSxtFoVS_5TKcjPFVXQd9C5LeTobap

# ===============================
# STREAMLIT APP CODE
# ===============================
app_code = """
import streamlit as st

st.set_page_config(page_title="ET Smart Life Concierge", layout="centered")

st.title("🚀 ET Smart Life Concierge")
st.subheader("AI-Powered Financial Assistant")

name = st.text_input("Enter your name")
age = st.number_input("Age", min_value=10, max_value=100, value=20)
income = st.number_input("Monthly Income (₹)", value=20000)
expenses = st.number_input("Monthly Expenses (₹)", value=10000)
goal = st.selectbox("Goal", ["save", "invest", "learn"])

def calculate_score(income, expenses):
    savings = income - expenses
    if savings <= 0:
        return 30, "Poor"
    elif savings < 5000:
        return 50, "Average"
    elif savings < 15000:
        return 75, "Good"
    else:
        return 90, "Excellent"

def finance_advice(income, expenses):
    savings = income - expenses
    if savings <= 0:
        return "Reduce unnecessary expenses immediately."
    elif income < 20000:
        return "Focus on saving and building emergency fund."
    elif income < 50000:
        return "Start SIP investments."
    else:
        return "Invest in Stocks + Mutual Funds."

def news(goal):
    if goal == "invest":
        return "Stock market trends & SIP strategies."
    elif goal == "save":
        return "Budgeting & saving tips."
    else:
        return "Learning & career growth."

def action(goal):
    if goal == "invest":
        return "Start SIP ₹1000/month."
    elif goal == "save":
        return "Save 20% income."
    else:
        return "Learn 1 hour daily."

if st.button("Generate AI Advice"):
    score, status = calculate_score(income, expenses)

    st.success("✅ AI Analysis Complete")
    st.write(f"👤 {name}")
    st.write(f"💡 Score: {score}/100 ({status})")
    st.write("💰", finance_advice(income, expenses))
    st.write("📰", news(goal))
    st.write("🚀", action(goal))
"""

# Save file
with open("app.py", "w") as f:
    f.write(app_code)

# ===============================
# RUN STREAMLIT IN BACKGROUND
# ===============================
def run():
    os.system("streamlit run app.py --server.port 8501")

threading.Thread(target=run).start()

# ===============================
# CREATE PUBLIC LINK
# ===============================
public_url = ngrok.connect(8501)
print("🚀 OPEN THIS LINK:", public_url)
```
